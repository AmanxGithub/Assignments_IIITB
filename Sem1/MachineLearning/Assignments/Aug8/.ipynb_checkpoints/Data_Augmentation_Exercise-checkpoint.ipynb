```python
import random
import skimage as sk
from skimage.transform import rescale, resize, downscale_local_mean
from skimage import util
import imageio
import warnings
import numpy 
from matplotlib.pyplot import imshow
import numpy as np
import cv2
from scipy import ndimage
warnings.filterwarnings("ignore")
```

```python
starry = cv2.imread('starry.jpeg')
starry_rgb = cv2.cvtColor(starry, cv2.COLOR_BGR2RGB)
imshow(starry_rgb)
```

```python

########### Gaussian Noise ##########

def noisy_gauss(img):
    mean = 0.0   
    std = 1.0
    noisy_img = cv2.add(img , np.random.normal(mean, std, img.shape).astype(np.uint8))   
    return noisy_img
```

```python
noisy_gauss_image =noisy_gauss(starry_rgb)
imshow(noisy_gauss_image)
```

```python

########### Salt and Pepper Noise ##########


def noisy_sp(img):
    s_vs_p = 0.5
    amount = 0.05
    out = np.copy(img)
    # Salt mode
    num_salt = np.ceil(amount * img.size * s_vs_p)
    coords = [np.random.randint(0, i - 1, int(num_salt))
            for i in img.shape]
    out[coords] = 1

    # Pepper mode
    num_pepper = np.ceil(amount* img.size * (1. - s_vs_p))
    coords = [np.random.randint(0, i - 1, int(num_pepper))
            for i in img.shape]
    out[coords] = 0
    return out
```

```python
noisy_sp_image =noisy_sp(starry_rgb)
imshow(noisy_sp_image)
```

```python
def find_coeffs(pa, pb):
        """ https://stackoverflow.com/questions/14177744/
            ????? ????????? ????, ??? ??????? - ???????
            """
        matrix = []
        for p1, p2 in zip(pa, pb):
            matrix.append([p1[0], p1[1], 1, 0, 0, 0, -p2[0]*p1[0], -p2[0]*p1[1]])
            matrix.append([0, 0, 0, p1[0], p1[1], 1, -p2[1]*p1[0], -p2[1]*p1[1]])
        A = numpy.matrix(matrix, dtype=numpy.float)
        B = numpy.array(pb).reshape(8)
        res = numpy.dot(numpy.linalg.inv(A.T * A) * A.T, B)
        return numpy.array(res).reshape(8)

def perspective_transform(img):
        width, height, ch = img.shape
        sym = img.copy()

        symmetry = False
        if random.uniform(0,1) < 1:
            from_points = [(0, 0), (width-1, 0), (width-1, height-1), (0, height-1)]
            new_points = [(width-1, 0), (0, 0), (0, height-1), (width-1, height-1)]
            coeffs = find_coeffs(new_points, from_points)
            sym = sym.transform((width, height), Image.PERSPECTIVE, coeffs, Image.BICUBIC)
            symmetry = True
        return sym, symmetry 
```
